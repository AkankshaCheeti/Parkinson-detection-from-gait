```python
import os
import pandas as pd
```

### Read the entries in dataset and analyze statistics

```python
# Set the path to the directory containing the files
path = '../dataset/'
count = 0
doc_len=[]
# Iterate through all the files in the directory
for filename in os.listdir(path):
    if filename.endswith('.txt'):  # Check if the file is a CSV file
        df = pd.read_csv(os.path.join(path, filename))
        doc_len.append(len(df))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Plot a histogram using Seaborn
sns.distplot(doc_len, kde=False)

# Set the x-axis label and title
plt.xlabel('count of datapoints')

plt.title('Histogram of count of datapoints')

# Show the plot
plt.show()
```

```python
import numpy as np
# Plot a histogram using Matplotlib
plt.hist(doc_len)

# Determine the value with the highest frequency
most_frequent_value = np.argmax(np.bincount(doc_len))

# Print the most frequent value
print('The most frequent value is:', most_frequent_value)

# Show the plot
plt.show()
```

```python
# Initialize the counters
count_5000 = 0
count_8000 = 0
count_10000 = 0
count_12000 = 0

# Loop through the list and update the counters
for num in doc_len:
    if num > 5000:
        count_5000 += 1
    if num > 8000:
        count_8000 += 1
    if num > 10000:
        count_10000 += 1
    if num > 12000:
        count_12000 += 1

# Print the results
print("Numbers greater than 5000:", count_5000)
print("Numbers greater than 8000:", count_8000)
print("Numbers greater than 10000:", count_10000)
print("Numbers greater than 12000:", count_12000)
```

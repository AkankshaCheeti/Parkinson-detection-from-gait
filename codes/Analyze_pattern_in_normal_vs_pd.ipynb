### Read data for normal vs pd and analyze each sensor and write conclusion

##### Time domain transformation

```python
file_path_nor = '../dataset/GaCo01_01.txt'
file_path_pd = '../dataset/GaPt03_01.txt'
cols=['time', 'l1','l2','l3','l4','l5','l6','l7','l8','r1','r2','r3','r4','r5','r6','r7','r8','total_l','total_r']
df_nor = pd.read_csv(file_path_nor, delimiter='\t', header=None, names=cols)
df_pd = pd.read_csv(file_path_pd, delimiter='\t', header=None, names=cols)

print(df_nor)
print(df_pd)
```

### Plot left foot data for Control 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create a list of the column names to loop over
cols = ['l1', 'l2', 'l3', 'l4', 'l5', 'l6', 'l7', 'l8']

# loop over each column and create a separate plot
for col in cols:
    fig, ax = plt.subplots(figsize=(10,5))
    sns.lineplot(data=df_nor, x='time', y=col, ax=ax)
    ax.set(title=f'Time series plot for {col}', xlabel='Time', ylabel='Value')
    plt.show()
```

### Plot left foot data for PD

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create a list of the column names to loop over
cols = ['l1', 'l2', 'l3', 'l4', 'l5', 'l6', 'l7', 'l8']

# loop over each column and create a separate plot
for col in cols:
    fig, ax = plt.subplots(figsize=(10,5))
    sns.lineplot(data=df_pd, x='time', y=col, ax=ax)
    ax.set(title=f'Time series plot for {col}', xlabel='Time', ylabel='Value')
    plt.show()
```
